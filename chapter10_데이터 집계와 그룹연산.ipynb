#### 데이터 그룹화와 집계
- groupby() : 데이터를 구분 할 수있는 열의 값들을 이용하여 데이터를 여러 기준에 의해 구분하여 그룹화 한 후 기초 통계 함수 등을 적용 할 수 있도록 함

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# species를 기준으로하여 iris의 그룹화
iris_grouped = iris.groupby(by=iris.species) # iris['species']와 같음
```

```python
iris_grouped
```

```python
# 그룹화 한것의 평균값 출력
iris_grouped.mean()
```

```python
# np.ravel() :  다차원 배열을 1차원 배열로 변환 시켜준다
iris['num'] = np.ravel([[i]*25 for i in range(6)]) # 0~5까지 숫자를 25개씩 생성, 2차원을 1차원으로 만듬
```

```python
iris
```

```python
iris_grouped2 = iris.groupby(by=[iris.species, iris.num])
```

```python
iris_grouped2.mean()
```

```python
# species를 기준으로 group화 한 데이터를 for문으로 확인하기
for key, group in iris_grouped:
    print(key, type(group), '\n', group.head())
```

```python
# 데이터프레임 그룹을 인덱싱 하는법
# 그룹을 인덱싱할 때 에는 take를 사용
iris_grouped.take([1, 2, 3])
```

```python
iris_grouped.take([0, 1, 2])
```

```python
# 슬라이싱은 불가능하다
# iris_grouped.take([:3, 1])
```

```python
# take를 사용해서만 인덱싱 할 수 있다
# iris_grouped.iloc[0,]
# iris_grouped.loc[0,]
```

```python
# get_group : by=(기준)을 주고 데이터를 가져오는 메소드
setosa_group = iris_grouped.get_group('setosa')
setosa_group
```

```python
setosa_group.shape
```

```python
setosa_group2 = iris_grouped2.get_group(('setosa',0)) 
# iris_grouped2의 by가 species와 num 2개로 생성되어 튜플로 묶어줌
```

### 와이드 포맷과 롱 포맷
- 와이드 포맷
  - 가로로 긴 형식의 데이터 구조
  - 열 단위 데이터 구조
  - 피벗 테이블 (pivot table)
- 롱 포맷
  - 세로로 긴 형식의 데이터 구조
  - 행 단위 데이터 구조
  - 언 피벗 테이블 (unpivot table)
- melt() : 와이드 포맷을 롱 포맷 데이터로 바꿈 (언피벗팅)
- pivot() : 롱 포맷을 와이드 포맷 데이터로 바꿈
